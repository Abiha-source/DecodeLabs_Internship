## 1.6 Dataset Visualisation

```python
# ── Task 1 Visualisation ─────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('#0d0d0d')
fig.suptitle('Task 1 — Dataset Overview: Product & Order Status Distribution',
             color=CYAN, fontsize=14, fontweight='bold')

# (a) Product Count — Horizontal Bar
prod_counts = df_raw['Product'].value_counts()
bars = axes[0].barh(prod_counts.index, prod_counts.values,
                     color=COLORS[:len(prod_counts)], edgecolor='#0d0d0d', linewidth=1.2)
axes[0].set_title('Product Distribution', color=CYAN, fontsize=12)
axes[0].set_xlabel('Number of Orders', color='white')
for bar, v in zip(bars, prod_counts.values):
    axes[0].text(v + 2, bar.get_y() + bar.get_height()/2,
                 str(v), va='center', color='white', fontsize=10)

# (b) Order Status — Pie Chart
status_counts = df_raw['OrderStatus'].value_counts()
axes[1].pie(status_counts.values, labels=status_counts.index,
            autopct='%1.1f%%', colors=COLORS[:len(status_counts)],
            wedgeprops=dict(edgecolor='#0d0d0d', linewidth=2),
            textprops={'color': 'white', 'fontsize': 9})
axes[1].set_title('Order Status Distribution', color=CYAN, fontsize=12)

plt.tight_layout()
plt.savefig('task1_overview.png', dpi=150, bbox_inches='tight', facecolor='#0d0d0d')
plt.show()
print("✅ Chart saved → task1_overview.png")
```
